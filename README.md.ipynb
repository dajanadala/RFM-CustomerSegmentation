### My recommendations
1.Focus on Best Customers and Loyal Customers:
"Best Customers" and "Loyal Customers" together make up over 50% of the customer base and have high average monetary values.
Recommendation: Develop targeted loyalty programs and exclusive offers to retain these high-value customers. Consider personalised marketing campaigns to enhance their shopping experience and incentivise continued purchases.
2. Re-engage Potential Loyalists and Promising Customers:
"Potential Loyalists" and "Promising" segments show a significant portion of the customer base but have lower monetary values and frequencies.
Recommendation: Implement engagement strategies such as personalised email campaigns, special discounts, and product recommendations to nurture these customers into becoming loyal customers. Monitor their activity and adjust strategies based on their responses.
3. Address At-Risk and Lost Customers:
The "At Risk" and "Lost" segments have high recency and low monetary values and frequencies, indicating they have not engaged recently.
Recommendation: Initiate reactivation campaigns to win back these customers. Offer attractive incentives, such as discounts or limited-time offers, and reach out to understand any issues they might have faced. Addressing their concerns can help regain their trust in business.

## Some of the screenshots from second task



### My recommendations
1. Focus on Early Engagement
The highest average revenue is generated within the first few weeks after registration, particularly in weeks 0 and 1.
Recommendation: Implement strategies to engage new users immediately after they register. This could include onboarding emails, special welcome discounts, or incentives for first-time purchases. Ensuring a positive and engaging initial experience can help maximise revenue from new users.
2. Identify and Reactivate Declining Cohorts
Revenue tends to decline significantly after the initial weeks, with some cohorts showing very low or even negligible revenue in later weeks.
Recommendation: Identify cohorts with sharp declines in revenue and target them with reactivation campaigns. Personalised emails, special offers, and reminders about products they showed interest in can help re-engage these users. Understanding the reasons for their decline through surveys or feedback can also provide valuable insights for improvement.
3. Optimize Marketing Spend Across Cohorts
Certain cohorts, such as those starting from 22/11/2020 and 29/11/2020, consistently show higher average revenues across multiple weeks compared to others.
Recommendation: Analyse the characteristics and acquisition channels of high-performing cohorts to replicate success. Allocate more marketing budget to channels and strategies that brought in these high-value cohorts. Additionally, study the engagement patterns of these cohorts to enhance retention strategies for other cohorts.
